# Segmentasi Pelanggan dengan RFM Analysis

```python
# Step 1: Hitung Recency
# Untuk menghitung Recency, kita perlu menentukan tanggal hari ini atau tanggal terakhir dalam dataset.
# Kita akan menggunakan tanggal terakhir dalam dataset df_Attribution sebagai tanggal referensi.
latest_date = df_Attribution['time'].max()

# Hitung jumlah hari sejak interaksi terakhir untuk setiap cookie
recency_df = df_Attribution.groupby('cookie')['time'].max().reset_index()
recency_df.columns = ['cookie', 'LastInteractionDate']
recency_df['Recency'] = (latest_date - recency_df['LastInteractionDate']).dt.days

print("Recency calculation results:")
display(recency_df.head())
```

```python
# Step 2: Hitung Frequency
# Hitung jumlah interaksi (baris) untuk setiap cookie
frequency_df = df_Attribution.groupby('cookie').size().reset_index(name='Frequency')

print("\nFrequency calculation results:")
display(frequency_df.head())
```

```python
# Step 3: Hitung Monetary
# Filter data hanya untuk interaksi 'conversion'
conversion_df = df_Attribution[df_Attribution['interaction'] == 'conversion'].copy()

# Hitung total conversion_value untuk setiap cookie
monetary_df = conversion_df.groupby('cookie')['conversion_value'].sum().reset_index()
monetary_df.columns = ['cookie', 'Monetary']

print("\nMonetary calculation results:")
display(monetary_df.head())
```

```python
# Step 4: Gabungkan RFM
# Gabungkan recency_df dan frequency_df
rfm_df = recency_df.merge(frequency_df, on='cookie', how='left')

# Gabungkan rfm_df dengan monetary_df
# Gunakan how='left' untuk memastikan semua cookie dari recency dan frequency disertakan,
# dan Monetary akan menjadi 0 jika tidak ada konversi.
rfm_df = rfm_df.merge(monetary_df, on='cookie', how='left').fillna(0)

# Hapus kolom LastInteractionDate karena tidak lagi diperlukan
rfm_df = rfm_df.drop('LastInteractionDate', axis=1)

print("\nCombined RFM DataFrame:")
display(rfm_df.head())
```

```python
# Step 5: Beri Skor RFM
# Untuk Recency, skor yang lebih rendah lebih baik (lebih baru)
# Untuk Frequency dan Monetary, skor yang lebih tinggi lebih baik (lebih sering/bernilai)

# Tentukan jumlah kuantil (misalnya, 4 untuk skor 1-4)
quantiles = rfm_df[['Recency', 'Frequency', 'Monetary']].quantile([.25, .5, .75]).to_dict()

# Fungsi untuk memberikan skor R
def r_score(x):
    if x <= quantiles['Recency'][0.25]:
        return 4
    elif x <= quantiles['Recency'][0.5]:
        return 3
    elif x <= quantiles['Recency'][0.75]:
        return 2
    else:
        return 1

# Fungsi untuk memberikan skor F dan M
def fm_score(x, c):
    if x <= quantiles[c][0.25]:
        return 1
    elif x <= quantiles[c][0.5]:
        return 2
    elif x <= quantiles[c][0.75]:
        return 3
    else:
        return 4

# Terapkan fungsi scoring ke DataFrame
rfm_df['R_score'] = rfm_df['Recency'].apply(lambda x: r_score(x))
rfm_df['F_score'] = rfm_df['Frequency'].apply(lambda x: fm_score(x, 'Frequency'))
rfm_df['M_score'] = rfm_df['Monetary'].apply(lambda x: fm_score(x, 'Monetary'))

print("\nRFM DataFrame with Scores:")
display(rfm_df.head())
```

```python
# Menjumlahkan skor R, F, dan M untuk membuat skor RFM numerik total
rfm_df['RFM_Sum_Score'] = rfm_df['R_score'] + rfm_df['F_score'] + rfm_df['M_score']

print("\nRFM DataFrame with RFM Sum Score:")
display(rfm_df.head())
```

```python
# ======================================
# Mapping Segmentasi 9 Kelompok RFM
# ======================================

def assign_segment(row):
    total = row['RFM_Sum_Score']   # total score (R+F+M)
    r = row['R_score']             # cek Recency Score

    if total == 9:
        return 'Champion'
    elif total == 8:
        return 'Loyal'
    elif total == 7:
        if r == 3:
            return 'Potential Loyalist'
        elif r == 1:
            return 'Cannot Lose Them'
        else:
            return 'Average'
    elif total == 6:
        if r == 3:
            return 'Recent Customer'
        elif r == 2:
            return 'Average'
        elif r == 1:
            return 'Cannot Lose Them'
        else:
            return 'Average'
    elif total in [4, 5]:
        return 'About To Sleep'
    elif total == 3:
        return 'Lost Customer'
    else:
        return 'Other'

# Buat kolom Segment
rfm_df['Segment'] = rfm_df.apply(assign_segment, axis=1)

# Lihat hasil
print(rfm_df[['cookie','R_score','F_score','M_score','RFM_Sum_Score','Segment']].head(10))

# Distribusi tiap segmen dalam bentuk tabel
print("\nDistribusi Segmentasi:")
segment_table = rfm_df['Segment'].value_counts().reset_index()
segment_table.columns = ['Segment', 'Jumlah_User']
display(segment_table)
```

```python
# Tabel detail per cookie
detail_table = rfm_df[['cookie','Recency','Frequency','Monetary',
                       'R_score','F_score','M_score','RFM_Sum_Score','Segment']]

display(detail_table.head(10))   # tampilkan 10 baris pertama
```

```python
# Ringkasan jumlah user tiap segmen
summary_table = rfm_df.groupby('Segment').agg(
    Jumlah_User=('cookie','count'),
    Avg_Recency=('Recency','mean'),
    Avg_Frequency=('Frequency','mean'),
    Avg_Monetary=('Monetary','mean')
).reset_index()

display(summary_table)
```
